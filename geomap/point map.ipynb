```python
def getDataOfDate(date):
    anger_l =  df[df['date'] == date]
    anger_l = anger_l[anger_l['emotion'] == 'anger'] 
   
    sadness_l =  df[df['date'] == date]
    sadness_l = sadness_l[sadness_l['emotion'] == 'sadness'] 
    
    joy_l =  df[df['date'] == date]
    joy_l = joy_l[joy_l['emotion'] == 'joy'] 
    
    fear_l =  df[df['date'] == date]
    fear_l = fear_l[fear_l['emotion'] == 'fear'] 
    
    disgust_l =  df[df['date'] == date]
    disgust_l = disgust_l[disgust_l['emotion'] == 'disgust'] 
    
    data = go.Data([
        go.Scattermapbox( # for anger
            lat = anger_l['latitude'],
            lon = anger_l['longitude'],
            mode = 'markers',
            marker = go.Marker(
                size = 7,
                color = 'rgb(255, 0, 0)',
                opacity = 0.8
            ),
            name = 'Anger',
            text = anger_l['tweet'] + anger_l['location'],
            hoverinfo = 'text'
        ),
        go.Scattermapbox( # for sadness
            lat = sadness_l['latitude'],
            lon = sadness_l['longitude'],
            mode = 'markers',
            marker = go.Marker(
                size = 7,
                color = 'rgb(0, 0, 255)',
                opacity = 0.8
            ),
            name = 'Sadness',
            text = sadness_l['tweet'] + sadness_l['location'],
            hoverinfo = 'text'
        ),
        go.Scattermapbox( # for joy
            lat = joy_l['latitude'],
            lon = joy_l['longitude'],
            mode = 'markers',
            marker = go.Marker(
                size = 7,
                color = 'rgb(255, 255, 0)',
                opacity = 0.8
            ),
            name = 'Joy',
            text = joy_l['tweet'] + joy_l['location'],
            hoverinfo = 'text'
        ),
        go.Scattermapbox( # for fear
            lat = fear_l['latitude'],
            lon = fear_l['longitude'],
            mode = 'markers',
            marker = go.Marker(
                size = 7,
                color = 'rgb(76, 0, 153)',
                opacity = 0.8
            ),
            name = 'Fear',
            text = fear_l['tweet'] + fear_l['location'],
            hoverinfo = 'text'
        ),
        go.Scattermapbox( # for disgust
            lat = disgust_l['latitude'],
            lon = disgust_l['longitude'],
            mode = 'markers',
            marker = go.Marker(
                size = 7,
                color = 'rgb(0, 255, 0)',
                opacity = 0.8
             ),
            name = 'Disgust',
            text = disgust_l['tweet'] + disgust_l['location'],
            hoverinfo = 'text'
        )
    ])

    return data
```
